```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
df = pd.read_csv('/content/covid_19_data.csv')
df.drop(['SNo','Last Update'],axis=1,inplace=True)
df.rename(columns={'ObservationDate':'Date','Province/State':'State','Country/Region':'Country'},inplace=True)
df['Date']=pd.to_datetime(df['Date'])
df.fillna('null',inplace=True)
df1 =df.groupby(['Country','Date'])[['Country','Date','Confirmed','Deaths','Recovered']].sum().reset_index()
df1.head(20)
```

```python
countries = df1['Country'].unique()
print(countries)
len(countries)
```

```python
for i in range(0,len(countries)):
  c = df1[df1['Country']==countries[i]].reset_index()
  plt.scatter(np.arange(0,len(c)),c['Confirmed'],color='blue',label='Confirmed')
  plt.scatter(np.arange(0,len(c)),c['Deaths'],color='red',label='Deaths')
  plt.scatter(np.arange(0,len(c)),c['Recovered'],color='green',label='Recovered')
  plt.title(countries[i])
  plt.xlabel('No of days since suspect')
  plt.ylabel('NO. of Cases')
  plt.legend()
  plt.show()
  
```

```python
df2=df1.groupby('Date')[['Country','Date','Confirmed','Deaths','Recovered']].sum().reset_index()
c=df2
plt.scatter(np.arange(0,len(c)),c['Confirmed'],color='blue',label='Confirmed')
plt.scatter(np.arange(0,len(c)),c['Deaths'],color='red',label='Deaths')
plt.scatter(np.arange(0,len(c)),c['Recovered'],color='green',label='Recovered')
plt.title('World')
plt.xlabel('No of days since suspect')
plt.ylabel('NO. of Cases')
plt.legend()
plt.show()
```
